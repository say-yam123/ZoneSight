```python
# Illegal Parking Hotspot Detection and Traffic Impact Analysis
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix
from scipy.stats import gaussian_kde
import warnings
warnings.filterwarnings('ignore')


print("=" * 80)
print("PARKING HOTSPOT DETECTION & TRAFFIC IMPACT MODELING")
print("=" * 80)

# Step 1: Create geographic grid for hotspot analysis
print("\nStep 1: Creating geographic grid for hotspot analysis...")

# Define grid resolution (0.005 degrees ≈ 500m in Bangalore)
grid_size = 0.005
df_cleaned['lat_grid'] = (df_cleaned['latitude'] // grid_size * grid_size).astype(str)
df_cleaned['lon_grid'] = (df_cleaned['longitude'] // grid_size * grid_size).astype(str)
df_cleaned['grid_cell'] = df_cleaned['lat_grid'] + "_" + df_cleaned['lon_grid']

print(f"  ✓ Created {df_cleaned['grid_cell'].nunique()} grid cells")


# Step 2: Calculate hotspot metrics at grid level
print("\nStep 2: Calculating hotspot metrics...")

grid_metrics = df_cleaned.groupby('grid_cell').agg({
    'id': 'count',  # Violation count
    'latitude': 'first',
    'longitude': 'first',
    'vehicle_type': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown',
    'police_station': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown'
}).rename(columns={'id': 'violation_count'})

grid_metrics['density'] = grid_metrics['violation_count']
grid_metrics = grid_metrics.sort_values('violation_count', ascending=False)

print(f"  ✓ Top 20 violation hotspots identified:")
print(grid_metrics.head(20)[['violation_count', 'vehicle_type', 'police_station']])


# Step 3: Feature engineering for traffic impact prediction
print("\nStep 3: Engineering features for traffic impact prediction...")

# Create time-based features
df_cleaned['hour'] = pd.to_datetime(df_cleaned['created_datetime']).dt.hour
df_cleaned['day_of_week'] = pd.to_datetime(df_cleaned['created_datetime']).dt.dayofweek
df_cleaned['is_peak_hour'] = df_cleaned['hour'].isin([7, 8, 9, 17, 18, 19]).astype(int)

# Merge grid metrics back to main dataframe
df_cleaned = df_cleaned.merge(
    grid_metrics[['violation_count', 'density']].rename(columns={'violation_count': 'cell_violations', 'density': 'cell_density'}),
    left_on='grid_cell',
    right_index=True,
    how='left'
)

# Create traffic impact features
df_cleaned['has_wrong_parking'] = df_cleaned['violation_type'].str.contains('WRONG PARKING', case=False).astype(int)
df_cleaned['has_no_parking'] = df_cleaned['violation_type'].str.contains('NO PARKING', case=False).astype(int)
df_cleaned['has_main_road'] = df_cleaned['violation_type'].str.contains('MAIN ROAD', case=False).astype(int)
df_cleaned['is_high_density_area'] = (df_cleaned['cell_density'] > df_cleaned['cell_density'].median()).astype(int)

# Get violation severity score
violation_severity = {
    'WRONG PARKING': 3,
    'NO PARKING': 3,
    'PARKING IN A MAIN ROAD': 5,
    'DOUBLE PARKING': 4,
    'PARKING ON FOOTPATH': 2,
    'PARKING NEAR ROAD CROSSING': 5,
    'PARKING NEAR BUSTOP/SCHOOL/HOSPITAL ETC': 4,
    'PARKING NEAR TRAFFIC LIGHT OR ZEBRA CROSS': 5
}

def calculate_violation_severity(violation_str):
    try:
        violations = json.loads(violation_str)
        severities = [violation_severity.get(v, 2) for v in violations]
        return max(severities) if severities else 2
    except:
        return 2

df_cleaned['violation_severity'] = df_cleaned['violation_type'].apply(calculate_violation_severity)

print(f"  ✓ Created {len(df_cleaned.columns)} features")
print(f"    - Traffic impact features: wrong_parking, no_parking, main_road")
print(f"    - Temporal features: hour, day_of_week, peak_hour")
print(f"    - Geographic features: cell_violations, cell_density, high_density_area")
print(f"    - Severity score: violation_severity (1-5)")


# Step 4: Prepare data for modeling
print("\nStep 4: Preparing data for modeling...")

# Remove rows with missing values in key features
model_data = df_cleaned[[
    'latitude', 'longitude', 'cell_violations', 'cell_density', 'hour', 
    'day_of_week', 'is_peak_hour', 'has_wrong_parking', 'has_no_parking', 
    'has_main_road', 'violation_severity', 'is_high_density_area'
]].dropna()

print(f"  ✓ Prepared {len(model_data)} records for modeling")
print(f"    - Training data shape: {model_data.shape}")


# Step 5: K-Means Clustering for Hotspot Detection
print("\nStep 5: Detecting hotspots using K-Means clustering...")

geo_features = df_cleaned[['latitude', 'longitude']].dropna()
scaler = StandardScaler()
geo_scaled = scaler.fit_transform(geo_features)

# Determine optimal number of clusters (elbow method)
inertias = []
K_range = range(3, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(geo_scaled)
    inertias.append(kmeans.inertia_)

# Use 6 clusters as optimal (can be adjusted)
optimal_k = 6
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_cleaned['hotspot_cluster'] = kmeans.fit_predict(geo_scaled)

print(f"  ✓ Identified {optimal_k} hotspot clusters")
hotspot_summary = df_cleaned.groupby('hotspot_cluster').agg({
    'id': 'count',
    'violation_severity': 'mean',
    'is_peak_hour': 'mean',
    'has_wrong_parking': 'sum'
}).rename(columns={'id': 'violation_count', 'is_peak_hour': 'peak_hour_ratio', 'has_wrong_parking': 'wrong_parking_count'})
print(hotspot_summary)

print("\n" + "=" * 80)
```

```python
# Model 1: Traffic Impact Prediction (Regression)
print("=" * 80)
print("MODEL 1: TRAFFIC IMPACT SEVERITY PREDICTION")
print("=" * 80)

# Prepare features and target
X = model_data[[
    'latitude', 'longitude', 'cell_violations', 'cell_density', 'hour',
    'day_of_week', 'is_peak_hour', 'has_wrong_parking', 'has_no_parking',
    'has_main_road'
]]

y_severity = model_data['violation_severity']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_severity, test_size=0.2, random_state=42)

# Train Gradient Boosting model for severity prediction
print("\nTraining Gradient Boosting Regressor for impact severity prediction...")
gbr_model = GradientBoostingRegressor(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    random_state=42
)
gbr_model.fit(X_train, y_train)

# Predictions
y_pred = gbr_model.predict(X_test)

# Evaluation
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"\n✓ Model trained successfully!")
print(f"  - R² Score: {r2:.4f}")
print(f"  - RMSE: {rmse:.4f}")
print(f"  - MSE: {mse:.4f}")

# Feature importance
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': gbr_model.feature_importances_
}).sort_values('Importance', ascending=False)

print(f"\nTop Features Affecting Traffic Impact:")
print(feature_importance.head(10).to_string(index=False))

# Model 2: High-Risk Zone Classification
print("\n" + "=" * 80)
print("MODEL 2: HIGH-RISK ZONE CLASSIFICATION")
print("=" * 80)

# Create binary target: high-risk zones (violation_severity >= 4)
y_risk = (model_data['violation_severity'] >= 4).astype(int)

X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X, y_risk, test_size=0.2, random_state=42
)

# Train Random Forest for high-risk zone classification
print("\nTraining Random Forest Classifier for high-risk zone detection...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train_rf, y_train_rf)

# Predictions
y_pred_rf = rf_model.predict(X_test_rf)
y_pred_proba = rf_model.predict_proba(X_test_rf)[:, 1]

# Evaluation
print(f"\n✓ Classification model trained successfully!")
print(f"\nClassification Report:")
print(classification_report(y_test_rf, y_pred_rf, target_names=['Low Risk', 'High Risk']))

# Feature importance for classification
feature_importance_rf = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print(f"\nTop Features for High-Risk Zone Detection:")
print(feature_importance_rf.head(10).to_string(index=False))

# Store predictions for later analysis
X_test['predicted_severity'] = y_pred
X_test['risk_probability'] = y_pred_proba
X_test['actual_severity'] = y_test.values

print("\n" + "=" * 80)
```

```python
# Targeted Enforcement Strategy & Recommendations
print("=" * 80)
print("TARGETED ENFORCEMENT STRATEGY")
print("=" * 80)

# Step 1: Rank hotspots by impact
print("\nStep 1: Ranking hotspots by traffic impact severity...")

hotspot_impact = df_cleaned.groupby('hotspot_cluster').agg({
    'id': 'count',
    'violation_severity': ['mean', 'max', 'sum'],
    'is_peak_hour': 'mean',
    'has_wrong_parking': 'sum',
    'has_no_parking': 'sum',
    'has_main_road': 'sum',
    'latitude': 'mean',
    'longitude': 'mean',
    'police_station': lambda x: x.mode()[0] if len(x.mode()) > 0 else 'Unknown'
}).round(2)

hotspot_impact.columns = ['violation_count', 'avg_severity', 'max_severity', 'total_severity',
                          'peak_hour_ratio', 'wrong_parking', 'no_parking', 'main_road',
                          'avg_latitude', 'avg_longitude', 'primary_station']

# Calculate impact score (weighted combination)
hotspot_impact['impact_score'] = (
    hotspot_impact['total_severity'] * 0.4 +
    hotspot_impact['violation_count'] * 0.3 +
    hotspot_impact['peak_hour_ratio'] * 100 * 0.2 +
    hotspot_impact['main_road'] * 0.1
)

hotspot_impact = hotspot_impact.sort_values('impact_score', ascending=False)

print("\nHotspot Ranking by Traffic Impact Score:")
print(hotspot_impact[['violation_count', 'avg_severity', 'peak_hour_ratio', 'impact_score',
                       'primary_station']].to_string())

# Step 2: Enforcement recommendations
print("\n\nStep 2: Enforcement recommendations by priority...")

enforcement_data = []
for idx, (cluster, row) in enumerate(hotspot_impact.iterrows(), 1):
    priority = 'CRITICAL' if idx <= 2 else 'HIGH' if idx <= 4 else 'MEDIUM'
    
    # Determine enforcement strategy
    if row['peak_hour_ratio'] > 0.4:
        enforcement_type = 'Peak-hour focused enforcement'
        timing = '7-9 AM, 5-7 PM'
    else:
        enforcement_type = 'All-day enforcement'
        timing = 'Round-the-clock'
    
    if row['main_road'] > row['violation_count'] * 0.15:
        focus = 'Main road & arterial streets'
    else:
        focus = 'Residential & market areas'
    
    enforcement_data.append({
        'Rank': idx,
        'Hotspot': f'Zone {cluster}',
        'Priority': priority,
        'Violations': int(row['violation_count']),
        'Avg_Severity': row['avg_severity'],
        'Impact_Score': row['impact_score'],
        'Station': row['primary_station'],
        'Strategy': enforcement_type,
        'Timing': timing,
        'Focus_Areas': focus
    })

enforcement_df = pd.DataFrame(enforcement_data)
print(enforcement_df.to_string(index=False))

# Step 3: Peak hour analysis
print("\n\nStep 3: Peak hour violation patterns...")

peak_analysis = df_cleaned[df_cleaned['is_peak_hour'] == 1].groupby(['hour', 'hotspot_cluster']).size().unstack(fill_value=0)
print(f"\nPeak Hour (7-9 AM, 5-7 PM) Violation Distribution:")
print(peak_analysis)

# Step 4: Quantify traffic impact
print("\n\nStep 4: Quantifying traffic impact on congestion...")

# Estimate congestion impact
total_violations = len(df_cleaned)
violations_main_road = (df_cleaned['has_main_road'] == 1).sum()
violations_peak_hour = (df_cleaned['is_peak_hour'] == 1).sum()
high_severity_violations = (df_cleaned['violation_severity'] >= 4).sum()

congestion_impact = pd.DataFrame({
    'Metric': [
        'Total Violations',
        'Main Road Violations (High Impact)',
        'Peak Hour Violations',
        'High Severity Violations',
        'Estimated Daily Congestion Hours (avg)',
        'Estimated Vehicles Affected'
    ],
    'Count': [
        total_violations,
        violations_main_road,
        violations_peak_hour,
        high_severity_violations,
        round(violations_main_road / 50),  # Estimate: 50 violations per congestion hour
        round(violations_main_road * 3)  # Estimate: 3 vehicles per violation
    ]
})

print(congestion_impact.to_string(index=False))

# Congestion severity estimation
avg_severity = df_cleaned['violation_severity'].mean()
print(f"\n\nAverage Violation Severity Score: {avg_severity:.2f}/5")
print(f"Estimated Congestion Impact:")
print(f"  - {violations_main_road:,} violations on main roads affecting traffic flow")
print(f"  - {violations_peak_hour:,} violations during peak hours ({(violations_peak_hour/total_violations*100):.1f}%)")
print(f"  - {high_severity_violations:,} high-severity violations requiring immediate action")

print("\n" + "=" * 80)
```
